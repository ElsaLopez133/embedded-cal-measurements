```python
import pandas as pd

```

```python
def extract_edges(df):
    pairs = []
    start = None
    for t, v in zip(df["Timestamp"], df["Value"]):
        if v == 1 and start is None:
            start = t
        elif v == 0 and start is not None:
            pairs.append((start, t))
            start = None
    return pairs


def range_stats(edge, df2):
    start, end = edge
    window = df2[(df2["Timestamp"] >= start) & (df2["Timestamp"] <= end)]
    return {
        "max": window["Value"].max(),
        "avg": window["Value"].mean(),
        "min": window["Value"].min(),
    }
```

```python
for edge in edges:
    cur = range_stats(edge, current)
    pwr = range_stats(edge, power)
    
    print(
        f"{(edge[1] - edge[0]):.6f}, "
        f"{cur['avg']}, {cur['max']}, {cur['min']}, "
        f"{pwr['avg']}, {pwr['max']}, {pwr['min']}"
    )
```
